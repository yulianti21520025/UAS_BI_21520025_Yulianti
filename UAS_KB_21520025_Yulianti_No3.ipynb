```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/caesarian.xlsx')
df
```

perlakuan dengan kondisi Ibu hamil dengan Usia 25 Tahun, yang 
merupakan Kelahiran ke -1, dengan Waktu kelahiran sesuai dengan HPL, Memiliki tekanan darah 
Normal

```python
#PERHITUNGAN DISTANCE
import math
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia'])-25)**4+ (float(df.iloc[i]['Kelahiran_ke-'])- 1)**4+
                         (float(df.iloc[i]['Waktu_Kelahiran'])-0)**4+ (float(df.iloc[i]['Tekanan_darah'])-1)**4))
    
```

```python
#perhitungan skala distance
df['dis']=dis
df    
```

```python
#sortir data berdasarkan jarak tetangga terdekat usia 25 tahun
x = df.sort_values('dis').head(5)
x
```

```python
#mengelompokkan data ID dan data Caesarian
y = x['Caesarian']
y
```

dari data diatas diperoleh bahwa perlakuaan kelahiran pada ID usia 25 tahun yang memiliki kondisi berbeda-beda adalah sebagai berikut :
ID ke 16 bernilai 0, yaitu Kelahiran Normal;
ID ke 46 bernilai 0, yaitu Kelahiran Normal;
ID ke 68 bernilai 1, yaitu Kelahiran Caesarian;
ID ke 55 bernilai 0, yaitu Kelahiran Normal; dan
ID ke 30  bernilai 0, yaitu Kelahiran Normal.

```python
#Menghitung rata-rata
np.mean(y)
```
