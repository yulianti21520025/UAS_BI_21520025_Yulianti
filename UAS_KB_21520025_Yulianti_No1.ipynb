```python
#Melakukan Tokeninsasi dengan model CountVectorizer untuk menghitung kata berdasarkan jumlah kemunculan

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
vectorizer_x = vectorizer.fit_transform(corpus).todense()
vectorizer_x
```

```python
#kata yang ada dalam dataset

vectorizer.get_feature_names()
```

```python
#menghitung jarak dokumen
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_x)):
  for j in range(i, len(vectorizer_x)):
    if i==j:
      continue
    jarak = euclidean_distances(vectorizer_x[i], vectorizer_x[j])
    print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')  
```
